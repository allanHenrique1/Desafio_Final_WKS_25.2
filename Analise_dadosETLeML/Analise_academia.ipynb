```python
import pandas as pd

df = pd.read_csv('academia_redfit.csv')

df['data_matricula'] = pd.to_datetime(df['data_matricula'], errors= 'coerce')
df['evolucao_PGC'] = df['primeiro_PGC'] - df['ultimo_PGC']

df['tipo_atividade'] = df['tipo_atividade'].replace({
    'Natacao' : 'Natação',
    'Swiming' : 'Natação',
    'Soccer' : 'Futebol',
    'Fut' : 'Futebol'
})

df = df.dropna()

df.to_csv("dados_tratados.csv", index=False)

df
```

```python
import matplotlib.pyplot as plt

# Contagem de cada categoria de sexo
contagem_sexo = df['sexo'].value_counts()

# Gráfico de barras
plt.figure(figsize=(6,4))
contagem_sexo.plot(kind='bar', color=['#1f77b4','#ff69b4','#9b59b6'], edgecolor='black')

plt.title("Distribuição de Sexo dos Alunos")
plt.xlabel("Sexo")
plt.ylabel("Quantidade de Alunos")
plt.xticks(rotation=0)
plt.show()
```

```python
# Gráfico de pizza
contagem = df['tipo_atividade'].value_counts()
plt.figure(figsize=(6,6))
contagem.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff9999'])

plt.title("Distribuição das Atividades")
plt.ylabel("")  # remove o label do eixo Y
plt.show()
```
